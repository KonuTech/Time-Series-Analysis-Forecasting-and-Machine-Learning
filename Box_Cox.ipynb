```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

```python
df = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Passengers'].plot(figsize=(12, 4))
```

```python
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
```

```python
df['SqrtPassengers'].plot(figsize=(12, 4))
```

```python
df['LogPassengers'] = np.log(df['Passengers'])
```

```python
df['LogPassengers'].plot(figsize=(12, 4))
```

```python
data, lam = boxcox(df['Passengers'])
```

```python
lam
```

```python
df['BoxCoxPassengers'] = data
```

```python
df['BoxCoxPassengers'].plot(figsize=(12, 4))
```
